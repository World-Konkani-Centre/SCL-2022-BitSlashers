#**Wheat** data Analysis

```python
df_wheat = pd.read_csv("/content/farmer-app-datasets/Datasets/Wheat Futures Historical Data (2).csv")
df_wheat= df_wheat.loc[::-1, :]
df_wheat['Price'] = df_wheat.apply(lambda row: 
  float(row['Price'].replace(",","")),
  axis=1
)
df_wheat.plot(x='Date', y='Price',title="Wheat",figsize=(20,10))
```
