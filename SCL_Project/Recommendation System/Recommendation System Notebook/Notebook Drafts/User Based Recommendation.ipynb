#Euclidean Distance Based Similarity finding


```python
def find_Euclidian_Similarity(Cust):
  df2 = df[df.columns.difference(["Names"])]
  cust_ind=df.index[df.Names==Cust]
  Cust_data=df2.values[cust_ind]
  dummy=np.full(shape=df.shape[0],fill_value=-1.0)
  df_sim_eu=pd.DataFrame(dummy,columns=["Euclidian Similarity"]);
  for ind in df2.index:
    if ind==cust_ind:
      continue;
    df_sim_eu["Euclidian Similarity"].values[ind]=spatial.distance.euclidean(Cust_data,df2.values[ind])
  df_sim_eu=df_sim_eu.copy()
  return df_sim_eu;
      
```

```python
df_sim_eu=find_Euclidian_Similarity("Johnny Palmer")
df_sim_eu.sort_values(by = "Euclidian Similarity",inplace=True)
print(df_sim_eu)
```

#Cosine Similarity

Cos(x, y) = x . y / ||x|| * ||y||

```python
def find_Cosine_Similarity(Cust):
  df2 = df[df.columns.difference(["Names"])]
  cust_ind=df.index[df.Names==Cust]
  Cust_data=df2.values[cust_ind]
  dummy=np.full(shape=df.shape[0],fill_value=-1.00)
  df_sim_co=pd.DataFrame(dummy,columns=["Cosine Similarity"]);
  for ind in df2.index:
    if ind==cust_ind:
      continue;
    df_sim_co["Cosine Similarity"].values[ind]=spatial.distance.cosine(Cust_data,df2.values[ind])
  return df_sim_co;
```

```python
df_sim_co=find_Cosine_Similarity("Johnny Palmer")
df_sim_co.sort_values(by = "Cosine Similarity",inplace=True)
print(df_sim_co)
```
