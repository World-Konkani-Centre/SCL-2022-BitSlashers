```python
import numpy as np
```

```python
d11=np.full(shape=100,fill_value="Wheat")
d21=np.random.randint(0,1000,100)
d31=np.random.randint(30,40,100)
```

```python
d12=np.full(shape=100,fill_value="Rubber")
d22=np.random.randint(0,1000,100)
d32=np.random.randint(10,20,100)
```

```python
d13=np.full(shape=100,fill_value="Barley")
d23=np.random.randint(0,1000,100)
d33=np.random.randint(60,70,100)
```

```python
d14=np.full(shape=100,fill_value="Maize")
d24=np.random.randint(0,1000,100)
d34=np.random.randint(100,110,100)
```

```python
d15=np.full(shape=100,fill_value="Rice")
d25=np.random.randint(0,1000,100)
d35=np.random.randint(30,40,100)
```

```python
d1=np.concatenate([d11,d12,d13,d14,d15],axis=None)
d2=np.concatenate([d21,d22,d23,d24,d25],axis=None)
d3=np.concatenate([d31,d32,d33,d34,d35],axis=None)
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(d1,columns=["Crops"]);
df2=pd.DataFrame(d2,columns=["Distance"]);
df3=pd.DataFrame(d3,columns=["Prices"]);
df=pd.concat([df1,df2,df3,names],axis=1);
print(df)
```

```python
df=df.sample(frac=1)
df.reset_index(drop=True, inplace=True)
print(df)
```

```python
df.to_csv('crop_data.csv', index=False)
```
