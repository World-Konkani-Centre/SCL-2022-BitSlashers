```python
import numpy as np
import pandas as pd
from scipy import spatial
```

```python
df=pd.read_csv("/content/farmer-app-datasets/Datasets/customer_data_ver_2.csv")
df.rename(columns={'FirstName LastName':"Names"},inplace=True)
# df.drop(columns=["index","index.1","index.2"],inplace=True)
# df=df.reset_index()
df
```

```python
def find_Cosine_Similarity(Cust):
  cust_ind=df.index[df.Names==Cust]
  ###Finding Similar Buyers
  df1=df[["Wheat Seller","Sugercane Seller","Rubber Seller","Maize Seller","Barley Seller"]]
  cust_data=df1.values[cust_ind]
  df_sim_buy=df1==cust_data
  df_sim_buy=df_sim_buy.replace(True,1).replace(False,0)
  ###Finding Similar Frequncy and Qunatity
  df2 = df[df.columns.difference(["Names","Wheat Seller","Sugercane Seller","Rubber Seller","Maize Seller","Barley Seller"])]
  Cust_data=df2.values[cust_ind]
  dummy=np.full(shape=df.shape[0],fill_value=-1.00)
  df_sim_co=pd.DataFrame(dummy,columns=["Cosine Similarity"]);
  for ind in df2.index:
    if ind==cust_ind:
      continue;
    df_sim_co["Cosine Similarity"].values[ind]=spatial.distance.cosine(Cust_data,df2.values[ind])
  df_sim=pd.concat([df_sim_co,df_sim_buy],axis=1)
  return df_sim;
```

```python
rand_ind=np.random.randint(0,499)
rand_name=df.loc[rand_ind].Names
df_sim=find_Cosine_Similarity(rand_name)
df_sim.sort_values(by = "Cosine Similarity",inplace=True)
# df_sim
i=0
indexes=[]
print("The Person to who the recommendation is done is "+rand_name)
print("The top 10 similar users to him/her are")
for ind in df_sim.index:
  if(i==11):
    break;
  indexes.append(ind)
  i=i+1
df.loc[indexes]
```
