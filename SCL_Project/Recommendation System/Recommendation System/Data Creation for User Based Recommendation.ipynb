```python
d_wheat=np.random.randint(1,10000,400)
d_paddy=np.random.randint(1,10000,400)
d_rubber=np.random.randint(1,10000,400)
d_maize=np.random.randint(1,10000,400)
d_na=np.full(shape=100,fill_value="NA")
```

```python
d_wheat=np.concatenate([d_wheat,d_na],axis=None)
d_paddy=np.concatenate([d_paddy,d_na],axis=None)
d_rubber=np.concatenate([d_rubber,d_na],axis=None)
d_maize=np.concatenate([d_maize,d_na],axis=None)
```

```python
np.random.shuffle(d_wheat)
np.random.shuffle(d_paddy)
np.random.shuffle(d_rubber)
np.random.shuffle(d_maize)
```

```python
df_wheat=pd.DataFrame(d_wheat,columns=["Wheat_A"]);
df_paddy=pd.DataFrame(d_paddy,columns=["Wheat_B"]);
df_rubber=pd.DataFrame(d_rubber,columns=["Wheat_C"]);
df_maize=pd.DataFrame(d_maize,columns=["Wheat_D"]);
df=pd.concat([df_name,df_wheat,df_paddy,df_rubber,df_maize],axis=1);
print(df)
```

```python
df.to_csv('customer_data.csv', index=False)
```
